```python
#Calculating the bounds
def calculate_bounds(df):
    bounds = {
        c: dict(
            zip(["q1", "q3"], df.approxQuantile(c, [0.25, 0.75], 0))
        )
        for c,d in zip(df.columns, df.dtypes) if d[1] == "int"
    }

    for c in bounds:
        iqr = bounds[c]['q3'] - bounds[c]['q1']
        bounds[c]['min'] = bounds[c]['q1'] - (iqr * 1.5)
        bounds[c]['max'] = bounds[c]['q3'] + (iqr * 1.5)

    return bounds
```

```python
#Flagging outliers
def flag_outliers(df, id_col):
  bounds = calculate_bounds(df)
  outliers = {}

  return df.select(c, id_col,
        *[
            f.when(
                ~f.col(c).between(bounds[c]['min'], bounds[c]['max']),
                "yes"
            ).otherwise("no").alias(c+'_outlier')
        ]
    )
```

```python
#GROUPING AGES go to [19, 20]
    #function to create group of ages, this helps because we have 43 differente values here
def Age(dataframe):
    dataframe.loc[dataframe['Age'] <= 30,'Age'] = 1
    dataframe.loc[(dataframe['Age'] > 30) & (dataframe['Age'] <= 40), 'Age'] = 2
    dataframe.loc[(dataframe['Age'] > 40) & (dataframe['Age'] <= 50), 'Age'] = 3
    dataframe.loc[(dataframe['Age'] > 50) & (dataframe['Age'] <= 60), 'Age'] = 4
    return dataframe

Age(pdf); 

sns.countplot(x = "Attrition",data=pdf,hue="Age")
plt.show()
```

```python
#GROUPING Monthly Income 
    #function to create group of monthly, this helps because we have 1,349 differente values here

def MonthlyIncome(dataframe):
     dataframe.loc[dataframe['MonthlyIncome'] <= 0,'MonthlyIncome'] = 1
     dataframe.loc[(dataframe['MonthlyIncome'] > 0) & (dataframe['MonthlyIncome'] <= 100000), 'MonthlyIncome'] = 2
     dataframe.loc[(dataframe['MonthlyIncome'] > 100000) & (dataframe['MonthlyIncome'] <= 150000), 'MonthlyIncome'] = 3
     dataframe.loc[(dataframe['MonthlyIncome'] > 150000) & (dataframe['MonthlyIncome'] <= 200000), 'MonthlyIncome'] = 4
     return dataframe
 
MonthlyIncome(pdf); 
 
sns.countplot(x = "Attrition",data=pdf,hue="MonthlyIncome")
plt.show()
```

```python
#GROUPING YearsWithCurrManager 
    #function to create five categories for this variable in order to make decision tree easier to interpret.

def YearsWithCurrManager(dataframe):
     dataframe.loc[dataframe['YearsWithCurrManager'] <= 1,'YearsWithCurrManager'] = 1
     dataframe.loc[(dataframe['YearsWithCurrManager'] > 1) & (dataframe['YearsWithCurrManager'] <= 5), 'YearsWithCurrManager'] = 2
     dataframe.loc[(dataframe['YearsWithCurrManager'] > 5) & (dataframe['YearsWithCurrManager'] <= 9), 'YearsWithCurrManager'] = 3
     dataframe.loc[(dataframe['YearsWithCurrManager'] > 9) & (dataframe['YearsWithCurrManager'] <= 13), 'YearsWithCurrManager'] = 4
     dataframe.loc[(dataframe['YearsWithCurrManager'] > 13) & (dataframe['YearsWithCurrManager'] <= 17), 'YearsWithCurrManager'] = 5
     return dataframe

YearsWithCurrManager(pdf); 
sns.countplot(x = "Attrition",data=pdf,hue="YearsWithCurrManager")
plt.show()
```

```python
#GROUPING YearsAtCompany
    #function to create six categories for this variable in order to make decision tree easier to interpret.

def YearsAtCompany(dataframe):
     dataframe.loc[dataframe['YearsAtCompany'] <= 1,'YearsAtCompany'] = 1
     dataframe.loc[(dataframe['YearsAtCompany'] > 1) & (dataframe['YearsAtCompany'] <= 3), 'YearsAtCompany'] = 2
     dataframe.loc[(dataframe['YearsAtCompany'] > 3) & (dataframe['YearsAtCompany'] <= 5), 'YearsAtCompany'] = 3
     dataframe.loc[(dataframe['YearsAtCompany'] > 5) & (dataframe['YearsAtCompany'] <= 10), 'YearsAtCompany'] = 4
     dataframe.loc[(dataframe['YearsAtCompany'] > 10) & (dataframe['YearsAtCompany'] <= 15), 'YearsAtCompany'] = 5
     dataframe.loc[(dataframe['YearsAtCompany'] > 15) & (dataframe['YearsAtCompany'] <= 20), 'YearsAtCompany'] = 6

     return dataframe
 
YearsAtCompany(winsPdf); 
 
sns.countplot(x = "Attrition",data=winsPdf,hue="YearsAtCompany")
plt.show()
```

```python
def ExtractFeatureImp(featureImp, dataset, featuresCol):
    list_extract = []
    for i in dataset.schema[featuresCol].metadata["ml_attr"]["attrs"]:
        list_extract = list_extract + dataset.schema[featuresCol].metadata["ml_attr"]["attrs"][i]
    varlist = pd.DataFrame(list_extract)
    varlist['score'] = varlist['idx'].apply(lambda x: featureImp[x])
    return(varlist.sort_values('score', ascending = False))
```

```python
#Other Forms of Interpretation
#total Results
lp = results.select('Attrition','prediction')
counttotal = results.count()
```

```python
#Correct
correct = lp.filter(lp['Attrition'] == lp['prediction']).count()
print("Correct:" + str(correct))
```

```python
#Wrong
wrong = lp.filter(~(lp['Attrition'] == lp['prediction'])).count()
print("Wrong:" + str(wrong))
```

```python
#TruePositives are how often the model correclty predicted a staff resigned
truep = lp.filter(lp['prediction'] == 0.0).filter(lp['Attrition'] == lp['prediction']).count()
print("TruePositive:" + str(truep))
```

```python
#FalseNegatives indicate how of the model preidcted a staff did not resign when in fact she or he did
falseN = lp.filter(lp['prediction'] == 0.0).filter(~(lp['Attrition'] == lp['prediction'])).count()
print("FalseNegative:" + str(falseN))
```

```python
#FalsePositives are how of the model predicted that staff resigned when in fact the staff did not
falseP = lp.filter(lp['prediction'] == 1.0).filter(~(lp['Attrition'] == lp['prediction'])).count()
print("FalsePositive:" +str(falseP))
```

```python
ratioWrong=wrong/counttotal
print("ratioWrong:" + str(ratioWrong))
ratioCorrect=correct/counttotal
print("ratioCorrect:" + str(ratioCorrect))
```

```python
#Tuning my model 
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

# Create ParamGrid for Cross Validation
paramGrid = (ParamGridBuilder()
             .addGrid(log_reg_attrition.regParam, [0.01, 0.5, 2.0])
             .addGrid(log_reg_attrition.elasticNetParam, [0.0, 0.5, 1.0])
             .addGrid(log_reg_attrition.maxIter, [1, 5, 10])
             .build())
```

```python
# Create 5-fold CrossValidator
cv = CrossValidator(estimator=log_reg_attrition, estimatorParamMaps=paramGrid, evaluator=my_eval, numFolds=5)

# Run cross validations
cvModel = cv.fit(train_attrition_data)
# this will likely take a fair amount of time because of the amount of models that we're creating and testing
```

```python
# Use test set to measure the accuracy of our model on new data
re_predictions = cvModel.transform(test_attrition_data)
```

```python
# cvModel uses the best model found from the Cross Validation
# Evaluate best model
my_eval.evaluate(re_predictions)
```

```python
#Access the model's feature weights and intercepts easily
print('Model Intercept: ', cvModel.bestModel.intercept)
```
